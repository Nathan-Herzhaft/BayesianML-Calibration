```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def Normal(µ,s) :
    def f(x) :
        Z = 1/(np.sqrt(2*math.pi) * s)
        exponent = -0.5*(((x - µ)/s)**2)
        return Z * np.exp(exponent)
    return f

f_1 = Normal(0,3)
f_2 = Normal(0,1)
```

```python
X = np.linspace(-5,5,500)
y_1 = f_1(X)
y_2 = f_2(X)
```

```python
fig, ax = plt.subplots(figsize = (8,8))
ax.plot(X,y_1,label = 'sigma = 3')
ax.plot(X,y_2,label = 'sigma = 1')
ax.axvline(0,color='grey',linewidth=0.5,label='µ')
ax.set_xlabel('y')
ax.set_ylabel('probability density')
ax.legend()
```
